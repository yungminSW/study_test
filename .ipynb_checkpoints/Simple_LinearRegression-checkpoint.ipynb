```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
# set the range of x
x = np.linspace(0,100, 100, dtype= np.float32)
# regression function
B_1 = 1.
B_0 = 3.6
h = B_1 * x + B_0
# random noise
mu,std = 0., 10.
e = np.random.normal(mu, std, x.shape)
# target value
y = h + e
# show result
plt.plot(x, h, label= 'regression function')
plt.plot(x, y, 'ro', alpha = 0.4, label = 'data point')
plt.grid()
plt.legend()
plt.xlabel("x")
plt.ylabel("y")
```

```python
# least squares estimator
x_c = np.subtract(x, np.mean(x))
y_c = np.subtract(y, np.mean(y))
square_norm_x_c = np.sum(np.power(x_c,2))
B_1_hat = np.dot(x_c,y_c) / square_norm_x_c
B_0_hat = np.mean(y) - B_1_hat * np.mean(x)
print('B_1_hat = ' , B_1_hat)
print('B_0_hat = ' , B_0_hat)
var_B_1_hat = std**2 / np.sum(np.power(x_c,2))
var_B_0_hat = std**2 / (1/x.shape[0] + np.mean(x)**2 / np.sum(np.power(x_c,2)))
print('VAR[B_1_hat] = ', var_B_1_hat)
print('VAR[B_0_hat] = ', var_B_0_hat)

# predicted value
y_hat = B_0_hat + B_1_hat * x

# coefficient of determination
SSE = np.sum(np.power(y - y_hat,2))
SSR = np.sum(np.power(y_hat - np.mean(y),2))
SST = SSE + SSR
R_2 = SSR / SST
print('Coefficient of determination = ', R_2)

# show result
plt.plot(x, h, 'b', label = 'regression function')
plt.plot(x, y_hat, 'g', label = 'predicted model')
plt.plot(x, y, 'ro', alpha = 0.4, label = 'data point')
plt.grid()
plt.legend()
plt.xlabel("x")
plt.ylabel("y")
```

```python
def t_distribution(dof):
  end = 10
  N = 1000
  t = np.linspace(-end, end, N)
  T = (math.gamma((dof+1)/2) / (math.gamma(dof/2)*math.sqrt(math.pi * dof))) \
      * np.power((1 + np.power(t,2)/dof), -(dof + 1) / 2.)
  return (2 * end / N),t, T

# get t-distribution 
delta, t, t_dist = t_distribution(x.shape[0])
# sample standard deviation
s = math.sqrt(np.sum(np.power(y - y_hat,2)) / (x.shape[0]-2))

''' B_1 = 0 test '''
# observed t value
O_t = B_1_hat / (s / math.sqrt(np.sum(np.power(x_c,2))))
print('observed value of t = ', O_t)
# get P-value
result = np.where(np.abs(t) > abs(O_t), t_dist, np.zeros_like(t))
P = np.sum(result) * delta
print('P-value = ', P)
# show result
plt.plot(t, t_dist, label = 't distribution')
plt.vlines(O_t,0. , 0.4,color="green",label = 'observed t')
plt.vlines(-O_t,0. , 0.4,color="green")
plt.grid()
plt.legend()
plt.title('B_1 test')
plt.xlabel("t")
plt.ylabel("p(t)")
```

```python
''' B_0 = 0 test '''
# observed t value
O_t = B_0_hat / (s *(1/x.shape[0] + np.mean(x)**2 / np.sum(np.power(x_c,2))))
print('observed value of t = ', O_t)
# get P-value
result = np.where(np.abs(t) > abs(O_t), t_dist, np.zeros_like(t))
P = np.sum(result) * delta
print('P-value = ', P)
# show result
plt.plot(t, t_dist, label = 't distribution')
plt.vlines(O_t,0. , 0.4,color="green",label = 'observed t')
plt.vlines(-O_t,0. , 0.4,color="green")
plt.grid()
plt.legend()
plt.title('B_0 test')
plt.xlabel("t")
plt.ylabel("p(t)")
```
